```python
#most lucky grounds for teams
def lucky(df,team_name):
    return matches[matches['winner']==team_name]['venue'].value_counts().nlargest(5)
```

```python
lucky(matches,'MI').plot(kind='bar')
```

```python
def comparison(team1,team2):
    compare=matches[((matches['team1']==team1)|(matches['team2']==team1))&((matches['team1']==team2)|(matches['team2']==team2))]
    sns.countplot(x='season', hue='winner',data=compare)
```

```python
comparison('MI','CSK')
```
